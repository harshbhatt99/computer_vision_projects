Basic Shapes

```python
canvas = np.zeros((300, 300, 3), dtype = "uint8")
# canvas.shape
# cv2.imshow("Canvas", canvas)
# cv2.waitKey(0)
plt.imshow(canvas, cmap='gray')
plt.show()
```

```python
# Opposite of canvas
white_canvas = cv2.bitwise_not(canvas)
plt.imshow(white_canvas, cmap='gray')
plt.show()
```

```python
# BGR Format (cv2)
# RGB Format (plt)
line_color = (0, 0, 255)
line_width = 5
rec_width = 10
cv2.line(white_canvas, (0, 0), (300, 300), line_color, line_width)
# cv2.imshow("Red Line on White Canvas", white_canvas)
# cv2.waitKey(0)
cv2.rectangle(white_canvas, (10, 10), (60, 60), line_color, rec_width)
# cv2.imshow("Red Square", white_canvas)
# cv2.waitKey(0)
plt.imshow(white_canvas, cmap='gray')
plt.show()
```

```python
canvas = np.zeros((300, 300, 3), dtype = "uint8")
(centerX, centerY) = (canvas.shape[1] // 2, canvas.shape[0] // 2)
white = (255, 255, 255)

for r in range(0, 175, 25):
    cv2.circle(canvas, (centerX, centerY), r, white)

# cv2.imshow("Circles Canvas", canvas)
# cv2.waitKey(0)
plt.imshow(canvas, cmap='gray')
plt.show()
```

```python
for i in range(0, 25):
    radius = np.random.randint(5, high = 200)
    color = np.random.randint(0, high = 256, size = (3,)).tolist()
    pt = np.random.randint(0, high = 300, size = (2,))
    cv2.circle(canvas, tuple(pt), radius, color, -1)

# cv2.imshow("Random Circles Canvas", canvas)
# cv2.waitKey(0)
plt.imshow(canvas, cmap='gray')
plt.show()
```

```python
# Create a black background
# If you don't put 3 here, the shape won't be changing the colors ...
# ... as there will be only one color channel

rectangle = np.zeros((300, 300, 3), dtype = "uint8")

# Create a rectangle with 2 corner co-ordinates, white color
# -1 means completely filled. This is for the thickness of rectangle

# cv2 >> BGR
# plt >> RGB

cv2.rectangle(rectangle, (25, 25), (275, 275), (125,52,47), -1)
# cv2.imshow("Rectangle", rectangle)
# cv2.waitKey(0)
plt.imshow(rectangle, cmap='gray')
plt.show()
```

```python
circle = np.zeros((300, 300, 3), dtype = "uint8")
cv2.circle(circle, (150, 150), 150, (10,250,150), -1)
# cv2.imshow("Circle", circle)
# cv2.waitKey(0)
plt.imshow(circle, cmap='gray')
plt.show()
```

```python
# Bitwise AND operation
# This will take the minimum comman pixels from both images
# So, upper-lower-side parts of circle and corners of rectangles will be eliminated
# It will take the minimum color possible

bitwiseAnd = cv2.bitwise_and(circle, rectangle)
# cv2.imshow("AND", bitwiseAnd)
# cv2.waitKey(0)
plt.imshow(bitwiseAnd, cmap='gray')
plt.show()
```

```python
# Bitwise OR operation
# Merge the shapes
# Higher color will be on top

bitwiseOr = cv2.bitwise_or(rectangle, circle)
# cv2.imshow("OR", bitwiseOr)
# cv2.waitKey(0)
plt.imshow(bitwiseOr, cmap='gray')
plt.show()
```

```python
# Bitwise XOR Operation
# Merge the objects with colors

bitwiseXor = cv2.bitwise_xor(rectangle, circle)
# cv2.imshow("XOR", bitwiseXor)
# cv2.waitKey(0)
plt.imshow(bitwiseXor, cmap='gray')
plt.show()
```

```python
# Bitwise NOT Operation

bitwiseNotCircle = cv2.bitwise_not(circle)
# cv2.imshow("NOT CIRCLE", bitwiseNotCircle)
plt.imshow(bitwiseNotCircle)
plt.show()

bitwiseNotRect = cv2.bitwise_not(rectangle)
# cv2.imshow("NOT RECT", bitwiseNotRect)
# cv2.waitKey(0)
plt.imshow(bitwiseNotRect, cmap='gray')
plt.show()
```

Math Operations on Image

```python
M = np.ones(image.shape, dtype = "uint8") * 100
plt.imshow(M, cmap='gray')
plt.show()
```

```python
# Increase the contrast of the image
added_image = cv2.add(image, M)
plt.imshow(added_image, cmap='gray')
plt.show()
```

```python
# Decrease the contrast of the image 
subtracted_image = cv2.subtract(image, M)
plt.imshow(subtracted_image, cmap='gray')
plt.show()
```

```python
(B, G, R) = cv2.split(image)
```

```python
plt.imshow(R, cmap='gray')
plt.show()
plt.imshow(G, cmap='gray')
plt.show()
plt.imshow(B, cmap='gray')
plt.show()
```

```python
merged = cv2.merge([B, G, R])
plt.imshow(merged, cmap='gray')
plt.show()
```

```python
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
cv2.imshow("Gray", gray)
cv2.waitKey(0)
```

```python
hsv = cv2.cvtColor(image, cv2.COLOR_BGR2HSV)
plt.imshow(hsv, cmap='gray')
plt.show()
```

```python
# LAB stands for Lightness, Green-Magenta, and Blue-Yellow as color space
lab = cv2.cvtColor(image, cv2.COLOR_BGR2LAB)
plt.imshow(lab, cmap='gray')
plt.show()
```

Image Manipulation

```python
cropped = image[30:120 , 240:335]
# cv2.imshow("Dexter", cropped)
# cv2.waitKey(0)
plt.imshow(cropped, cmap='gray')
plt.show()
```

```python
# 1 (axis) >> Horizontal flit
h_flipped = cv2.flip(image, 1)
plt.imshow(h_flipped, cmap='gray')
plt.show()
```

```python
# 0 (axis) >> Vertical Flip
v_flipped = cv2.flip(image, 0)
plt.imshow(v_flipped, cmap='gray')
plt.show()
```

```python
# -1 (axis) >> Horizontal + Vertical Flip
hv_flipped = cv2.flip(image, -1)
plt.imshow(hv_flipped, cmap='gray')
plt.show()
```

```python
# Create a black mask with the size of image
# :2 retrieves first 2 elements from the shape which is width and height (not channels)
mask = np.zeros(image.shape[:2], dtype = "uint8")

# Finding center of the image >> width / 2, height / 2
(cX, cY) = (image.shape[1] // 2, image.shape[0] // 2)

# Create a rectangle based on mask
cv2.rectangle(mask, (cX - 75, cY - 75), (cX + 75 , cY + 75), 255, -1)

# Circular mask
# cv2.circle(mask, (cX, cY), 100, 255, -1)

# Create white square inside black background
cv2.imshow("NewMask", mask)
cv2.waitKey(0)

# plt color channel is different so the below output is different than cv2
plt.imshow(mask, cmap='gray')
plt.show()
```

```python
masked = cv2.bitwise_and(image, image, mask = mask)

cv2.imshow("Mask Applied to Image", masked)
cv2.waitKey(0)

plt.imshow(masked, cmap='gray')
plt.show()
```
